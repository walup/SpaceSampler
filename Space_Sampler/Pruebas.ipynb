```python
#Importamos la clase implementada para muestrear puntos de un espacio
from SpaceSampler import SpaceSampler
import matplotlib.pyplot as plt
#Importamos los tipos de distancia, hasta ahora solo hay Euclidiana, Wasserstein y Aitchinson
from DistanceType import DistanceType
import numpy as np
```

#### Ejemplificaci√≥n de uso del muestreo de puntos

#### Distancia euclidiana
##### Bola

```python
#Primero como ejemplo obtengamos puntos en una bola centrada en [3,3] y con radio 3
center = [3,3]
radius = 3
nPoints = 1000
distanceType = DistanceType.EUCLIDEAN_DISTANCE
spSamp = SpaceSampler()


ballPoints = spSamp.sampleBallPoints(center,radius,nPoints, distanceType)
```

```python
#Vamos a graficar
plt.figure()
plt.plot(ballPoints[:,0], ballPoints[:,1], color = "#2fcaf5", marker = "o", linestyle = "none")
plt.axis('equal');
```

##### Anillo

```python
#Ahora vamos a obtener los puntos de un anillo cuyo radio interior es 1 y radio exterior es 3
center = [3,3]
radius1 = 1
radius2 = 3
distanceType = DistanceType.EUCLIDEAN_DISTANCE

annulusPoints = spSamp.sampleAnnulusPoints(center, radius1, radius2, nPoints, distanceType)
```

```python
plt.figure()
plt.plot(annulusPoints[:,0], annulusPoints[:,1], color = "#2fcaf5", marker = "o", linestyle = "none")
plt.axis('equal');
```

```python
center = [3,3]
radius = 3
nPoints = 1000
distanceType = DistanceType.EUCLIDEAN_DISTANCE
spSamp = SpaceSampler()


perimeterPoints = spSamp.samplePerimeterPoints(center,radius,nPoints,distanceType)
```
