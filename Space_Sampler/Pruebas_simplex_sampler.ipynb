```python

#Importamos la clase implementada para muestrear puntos de un espacio
from SimplexSpaceSampler import SimplexSpaceSampler
import matplotlib.pyplot as plt
#Importamos los tipos de distancia, hasta ahora solo hay Euclidiana, Wasserstein y Aitchinson
from DistanceType import DistanceType
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
```

#### Euclidiana

```python
center = np.array([1,1,1])
radius = 0.1
nPoints = 1000
distanceType = DistanceType.EUCLIDEAN_DISTANCE
spSamp = SimplexSpaceSampler()


ballPoints = spSamp.sampleBallPoints(center,radius,nPoints, distanceType)
```

```python
simplexPoints = spSamp.sampleSimplexPointsOutsideCircle(center,radius,nPoints,distanceType)
fig = plt.figure(figsize = (10, 7))
ax = fig.add_subplot(projection='3d')
# Creating plot
ax.scatter3D(ballPoints[:,0], ballPoints[:,1], ballPoints[:,2], color = "#fadf46", s= 0.5)
ax.scatter3D(simplexPoints[:,0], simplexPoints[:,1], simplexPoints[:,2], color = "#0071bd",s = 0.5)
plt.title("Euclidiana - Bola")
ax.view_init(25,-26)
plt.show()
```
